```python
import matplotlib.pyplot as plt
```

```python
# LINE CHART
week_days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
margherita_pizza_sold = [270, 510, 310, 710, 830, 990, 730]

# week days on x-axis and pizza sold on y-axis
plt.plot(week_days, margherita_pizza_sold, marker='o', linestyle='solid')
plt.title("Margherita pizza sales") # add title
plt.ylabel("Amount (£)") # add a lable to y-axis
plt.show() # display with show() or save it with savefig()
```

```python
# LINE CHART
week_days = [1, 2, 3, 4, 5, 6, 7]
sales = [['Margherita', 270], ['Margherita', 510], ['Margherita', 310], ['Margherita', 710], ['Margherita', 830], 
         ['Margherita', 990], ['Margherita', 730], 
         ['Pepperoni', 210], ['Pepperoni', 300], ['Pepperoni', 440], ['Pepperoni', 600], ['Pepperoni', 900], 
         ['Pepperoni', 1120], ['Pepperoni', 500]]

m = [sale[1] for sale in sales if sale[0] == 'Margherita']
p = [sale[1] for sale in sales if sale[0] == 'Pepperoni']

plt.plot(week_days, m, color='r', marker='o', label='Margherita')
plt.plot(week_days, p, color='g', marker='o', label='Pepperoni')
plt.xticks([n for n in week_days] ,['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], rotation=45) # add x labels
plt.ylabel("Amount (£)") # add a lable to y-axis
plt.legend(loc=9) # add legend
plt.title("Pizzas sales") # add title
plt.show()
```

```python
# BAR CHART
from collections import Counter # dict subclass for counting objects

java_grades = [83, 95, 91, 87, 70, 61, 85, 82, 55, 100, 67, 41, 73, 77, 0, 49]
decile = lambda grade: (grade // 10) * 10 # // stands for int division
histogram = Counter(decile(grade) for grade in java_grades) # returns dict(key : value) - grade : # of students

plt.bar([x for x in histogram.keys()],     
        histogram.values(),                # give each bar its height
        8)                                 # third argument is the bar width
plt.axis([-5, 105, 0, 5])                  # x-axis from -5 to 105, and y-axis from 0 to 5
plt.xticks([10 * i for i in range(11)])    # x-axis labels at 0, 10, 20, ..., 100
plt.xlabel("Decile grade")
plt.ylabel("# of Students")
plt.title("Distribution of Java exam grades")
plt.show()
```

```python
#SCATTER PLOT
kds = [7.0, 6.5, 7.2, 4.3, 7.1, 3.4, 2.0, 5.4, 6.7]
hours = [175, 170, 205, 120, 220, 130, 105, 145, 190]
players = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(kds, hours)
# label each point
for player, kd, hours_count in zip(players, kds, hours):
    plt.annotate(player,
                 xy=(kd, hours_count), # put the label with its point
                 xytext=(5, -5), # but slightly offset
                 textcoords='offset points')
plt.title("K.D. vs Number hours playing the game")
plt.xlabel("K.D.")
plt.ylabel("# of hours")
plt.show()
```

```python
# SCATTER PLOT
# if you are scattering comparable variables
# you can get a misleading viz if you let matplotlib choose the scale
minutes = [99, 90, 85, 97, 80]
secrets = [100, 85, 60, 90, 70]

plt.scatter(minutes, secrets)
plt.title("Axes aren't comparable")
plt.xlabel("# of minutes")
plt.ylabel("Secrets discovered")
plt.show()
```

```python
# if we include a call to plt.axis("equal")
# the plot shows more accurately the variation
plt.scatter(minutes, secrets)
plt.title("Axes are comparable")
plt.axis("equal")
plt.xlabel("# of minutes")
plt.ylabel("Secrets discovered")
plt.show()
```
